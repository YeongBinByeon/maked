```python

def compute_icon_type(img):
    mean = np.mean(img, axis=(0,1))
    result = 'gjgi'
    
    if mean[0] > 50 and mean[0] < 55 and mean[1] > 50 and mean[1] < 55 and mean[2] > 50 and mean[2] < 55:
        result = 'BOMB'
    elif mean[0] > 250 and mean[1] > 85 and mean[1] < 110 and mean[2] > 250:
        result = 'SWORD'
    elif mean[0] > 100 and mean[0] < 130 and mean[1] > 150 and mean[1] < 200 and mean[2] > 90 and mean[2] < 110:
        result = 'POISON'
    elif mean[0] > 210 and mean[0] < 230 and mean[1] > 200 and mean[1] < 225 and mean[2] > 120 and mean[2] < 135:
        result = 'JEWEL'
    return result


def compute_icon_type_(img):
    mean = np.mean(img, axis=(0,1))
    result = 'gjgi'
    
    if mean[0] > 180 and mean[0] < 190 and mean[1] > 130 and mean[1] < 140 and mean[2] > 30 and mean[2] < 40:
        result = 'FIRST'
    elif mean[0] > 80 and mean[0] < 90 and mean[1] > 100 and mean[1] < 110 and mean[2] > 150 and mean[2] < 160:
        result = 'SECOND'
    elif mean[0] > 90 and mean[0] < 110 and mean[1] > 90 and mean[1] < 110 and mean[2] > 90 and mean[2] < 110:
        result = 'THIRD'
    elif mean[0] > 210 and mean[0] < 230 and mean[1] > 200 and mean[1] < 225 and mean[2] > 120 and mean[2] < 135:
        result = 'JEWEL'
    return result

def click(coords):
    pag.moveTo(x=coords[0], y=coords[1], duration=0.0)
    pag.mouseDown()
    pag.mouseUp()
    
```
